```python
import pandas as pd

df = pd.read_csv("company_metrics.csv")

lista_symbols = df['Symbol'].unique().tolist()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np



data = []
errores = []  

for sym in lista_symbols:
    try:
        ticker = yf.Ticker(sym)
        info = ticker.info

        # Si no tiene info, consideramos que ha fallado
        if not info or "regularMarketPrice" not in info:
            errores.append(sym)
            continue

        # --- Extraemos métricas directamente si están disponibles
        current_ratio = info.get("currentRatio")
        quick_ratio = info.get("quickRatio")
        debt_to_equity = info.get("debtToEquity")
        interest_coverage = info.get("interestCoverage")
        fcf = info.get("freeCashflow")
        ebitda = info.get("ebitda")
        trailing_pe = info.get("trailingPE")
        peg_ratio = info.get("pegRatio")

        # --- Cálculo de crecimiento medio de ingresos 5 años ---
        try:
            fin = ticker.financials
            rev = fin.loc["Total Revenue"]
            if len(rev) >= 2:
                rev_growth = (rev.iloc[0] / rev.iloc[-1]) ** (1 / (len(rev) - 1)) - 1
            else:
                rev_growth = np.nan
        except Exception:
            rev_growth = np.nan

        # --- Guardamos resultados ---
        data.append({
            "symbol": sym,
            "Current Ratio": current_ratio,
            "Quick Ratio": quick_ratio,
            "Debt/Equity": debt_to_equity,
            "Interest Coverage": interest_coverage,
            "Free Cash Flow": fcf,
            "5Y Avg Revenue Growth": rev_growth,
            "EBITDA": ebitda,
            "PER (Trailing)": trailing_pe,
            "PEG Ratio": peg_ratio
        })
    except Exception as e:
        print(f"❌ Error procesando {sym}: {e}")
        errores.append(sym)

# -------------------------------
# 📊 RESULTADOS
# -------------------------------
df = pd.DataFrame(data)
pd.set_option("display.float_format", "{:,.2f}".format)
display(df)

# df.to_csv("metricas_empresas.csv", index=False)

# -------------------------------
# ⚠️ LISTA DE SYMBOLS NO ENCONTRADOS
# -------------------------------
if errores:
    print("\n❌ No se pudieron obtener datos para los siguientes símbolos:")
    print(errores)
else:
    print("\n✅ Todos los símbolos se procesaron correctamente.")
```
