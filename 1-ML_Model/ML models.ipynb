  - **Valores a probar:** `2`, `3`

**5. MarkovBlanketClassifier**

- **Par√°metro:** `search`

  - **Valores a probar:** `BestFirst (Default)`, `GreedyStepwise`


**6. BayesNet ‚Äî Red Bayesiana General y TAN**

- **Par√°metro:** `searchAlgorithm`
